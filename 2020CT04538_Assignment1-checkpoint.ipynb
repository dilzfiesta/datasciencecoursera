This dataset consists of 303 observations of Patients. 

There are 11 features or attributes with various values and description:
* Age : Age of the patients,                       
* Sex : 1- Male, 0 - Female , 
* CP_Type : Chest pain type as 0 = typical angina, 1 = atypical angina, 2 = non-anginal pain,  3 = asymptomatic ,
* Blood Pressure : Blood pressure values of patients , 
* Cholestrol : Cholestrol values of patients , 
* BloodSugar : Patient has blood sugar 1 = YES, 0 = NO , 
* ECG : ECG Values 0 = Nothing to note, 1 = ST-T Wave abnormality, 2 = Possible or definite left ventricular hypertrophy,
* MaxHeartRate : Patients Maximum Heart Rate values,
* ExerciseAngia : Patient having Exercise Angia  1 = YES, 0 = NO,
* FamilyHistory : Number of persons affected in the family
* Target  : 0 = LESS CHANCE , 1 = MORE CHANCE


The purpose for this dataset is for 
* Building a Predictive Model using Na√Øve Bayesian Approach (Which features decide heart attack?)
* Evaluate the model using AUC-ROC, Precision, Recall, F_score, Accuracy
